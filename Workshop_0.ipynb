## 4.1 Exercise on Functions


### Task - 1:

Create a Python program that converts between different units of measurement.

• The program should:

1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).  
2. Ask the user to input the value to be converted.  
3. Perform the conversion and display the result.  
4. Handle potential errors, such as invalid input or unsupported conversion types.  

• Requirements:

1. Functions: Define at least one function to perform the conversion.  
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).  
3. User Input: Prompt the user to select the conversion type and input the value.  
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and return value.  

• Conversion Options:

1. Length:  
– Convert meters (m) to feet (ft).  
– Convert feet (ft) to meters (m).  

2. Weight:  
– Convert kilograms (kg) to pounds (lbs).  
– Convert pounds (lbs) to kilograms (kg).  

3. Volume:  
– Convert liters (L) to gallons (gal).  
– Convert gallons (gal) to liters (L).  


```python
def convert_units(conversion_type: str, value: float) -> float:
    """
    Convert a value between supported measurement units.

    Args:
        conversion_type (str): Type of conversion
            (m_to_ft, ft_to_m, kg_to_lbs, lbs_to_kg, l_to_gal, gal_to_l)
        value (float): Numeric value to convert

    Returns:
        float: Converted value
    """
    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type")


print("Choose conversion type:")
print("1. m_to_ft")
print("2. ft_to_m")
print("3. kg_to_lbs")
print("4. lbs_to_kg")
print("5. l_to_gal")
print("6. gal_to_l")

try:
    choice = input("Enter conversion type: ").strip()
    value = float(input("Enter value: ").strip())

    result = convert_units(choice, value)
    print(f"Converted value: {result:.4f}")

except ValueError as e:
    print(f"Error: {e}")
```

## Task - 2:

Create a Python program that performs various mathematical operations on a list of numbers.

• The Program should:

1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).

3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists.

• Requirements:

1. Functions: Define at least one function for each operation (sum, average, maximum, minimum).

2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or empty lists).

3. User Input: Prompt the user to select the operation and input the list of numbers.

4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and return value.


```python
def perform_operations(operation_type: int, numbers: list) -> float:
    """
    Calculates the sum, average, maximum or minimum of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The mathematical operation of the numbers.
    """
    if operation_type == 1:
      return sum(numbers)
    elif operation_type == 2:
      return sum(numbers) / len(numbers)
    elif operation_type == 3:
      return max(numbers)
    elif operation_type == 4:
      return min(numbers)


print("Choose an operation:")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

try:
    choice = input("Enter operation choice (1-4): ").strip()
    if choice not in ["1", "2", "3", "4"]:
        raise ValueError("Invalid operation choice")

    numbers_input = input("Enter a list of numbers separated by spaces: ").strip()

    if not numbers_input:
        raise ValueError("Input list cannot be empty.")

    numbers = [float(num) for num in numbers_input.split()]

    print(perform_operations(int(choice), numbers))

except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## 4.2 Exercise on List Manipulation


```python
def extract_every_other(lst):
    """Returns every other element starting from the first."""
    return lst[::2]

def get_sublist(lst, start, end):
    """Returns sublist from start to end (inclusive)."""
    return lst[start:end+1]

def reverse_list(lst):
    """Reverses a list using slicing."""
    return lst[::-1]

def remove_first_last(lst):
    """Removes first and last elements."""
    return lst[1:-1]

def get_first_n(lst, n):
    """Extracts first n elements."""
    return lst[:n]

def get_last_n(lst, n):
    """Extracts last n elements."""
    return lst[-n:]

def reverse_skip(lst):
    """Reverse order starting from second-to-last, skipping one."""
    return lst[-2::-2]

# --- Execution Tests ---
my_list = [1, 2, 3, 4, 5, 6]
print("Every Other:", extract_every_other(my_list))   # Output: [1, 3, 5]
print("Sublist (2-4):", get_sublist(my_list, 2, 4)) # Output: [3, 4, 5]
print("Reverse Skip:", reverse_skip(my_list))        # Output: [5, 3, 1]
```

4.3 Exercise on Nested List

```python
def flatten(lst):
    """Flattens a 2D list into 1D."""
    return [item for sublist in lst for item in sublist]

def access_nested_element(lst, indices):
    """Accesses element using a list of indices."""
    for i in indices:
        lst = lst[i]
    return lst

def sum_nested(lst):
    """Recursive sum of all numbers in any nested depth."""
    total = 0
    for i in lst:
        total += sum_nested(i) if isinstance(i, list) else i
    return total

def remove_element(lst, elem):
    """Removes all occurrences of elem from nested list."""
    return [remove_element(i, elem) if isinstance(i, list) else i for i in lst if i != elem]

def find_max(lst):
    """Finds maximum value in a deeply nested list."""
    def get_all(l):
        for i in l:
            yield from get_all(i) if isinstance(i, list) else [i]
    return max(list(get_all(lst)))

def deep_flatten(lst):
    """Flattens list of any depth."""
    res = []
    for i in lst:
        res.extend(deep_flatten(i)) if isinstance(i, list) else res.append(i)
    return res

def average_nested(lst):
    """Calculates average of all elements in nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

nested = [[1, 2], [3, [4, 5]], 6]
print("Flattened:", deep_flatten(nested))
print("Total Sum:", sum_nested(nested))
print("Max Value:", find_max(nested))
```
